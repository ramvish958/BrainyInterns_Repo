# "The Economic_Indicators_dataset covers the years from 2010 to 2023."

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("economic_indicators_dataset_2010_2023.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data['Country'].unique()
```

```python
Groupby_GDP_Inflation = data.groupby("Country")[["GDP Growth Rate (%)","Inflation Rate (%)"]].mean()
```

```python

plt.figure(figsize=(8,5))
plt.title('GDP Growth Rate and Inflation Rate Over Country')

plt.plot(Groupby_GDP_Inflation.index, Groupby_GDP_Inflation["GDP Growth Rate (%)"], 
         color='blue', marker='o', linestyle='-', label='GDP Growth Rate (%)')

plt.plot(Groupby_GDP_Inflation.index, Groupby_GDP_Inflation["Inflation Rate (%)"], 
         color='red', marker='s', linestyle='--', label='Inflation Rate (%)')
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 1))
plt.grid(True)
plt.xlabel("Country")
plt.ylabel("GDP_Inflation rate (%)")
plt.show()
```

```python
Groupby_GDP_Unimployment = data.groupby("Country")[["GDP Growth Rate (%)","Unemployment Rate (%)"]].mean()
```

```python
plt.figure(figsize=(8,5))
plt.title('GDP Growth Rate and Unimployment Rate Over Country')
plt.plot(Groupby_GDP_Unimployment.index,Groupby_GDP_Unimployment["GDP Growth Rate (%)"],
         color="blue",marker="o",linestyle="-",label='GDP Growth Rate (%)')
plt.plot(Groupby_GDP_Unimployment.index,Groupby_GDP_Unimployment["Unemployment Rate (%)"],
         color="yellow",marker="p",linestyle="--",label="Unemployment Rate (%)")
plt.xlabel("Country")
plt.ylabel("GDP_Unimployment rate (%)")
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 1))
plt.grid(True)
plt.show()
```

```python
Group_by_GDP_Inflation_unemployement = data.groupby("Country")[["GDP Growth Rate (%)","Inflation Rate (%)","Unemployment Rate (%)"]].mean()
```

```python
plt.figure(figsize=(8,5))
plt.plot(Group_by_GDP_Inflation_unemployement.index,Group_by_GDP_Inflation_unemployement["GDP Growth Rate (%)"],
        color="blue",marker="o",linestyle="-",label="GDP Growth Rate (%)")
plt.plot(Group_by_GDP_Inflation_unemployement.index,Group_by_GDP_Inflation_unemployement["Inflation Rate (%)"],
        color="red",marker="s",linestyle="--",label="Inflation Rate (%)")
plt.plot(Group_by_GDP_Inflation_unemployement.index,Group_by_GDP_Inflation_unemployement["Unemployment Rate (%)"],
        color="yellow",marker="h",linestyle="--",label="Unemployment Rate (%)")
plt.xlabel("Country")
plt.ylabel("GDP_Inflation_Unimployement rate (%)")
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 1))
plt.grid(True)
plt.show()
```
