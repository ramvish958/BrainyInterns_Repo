# **We are using "SuperStore dataset" for pivot and pivot_table Operation**

```python
import pandas as pd
```

```python
data = pd.read_csv("SuperStore.csv")
```

```python
data.head()
```

```python
grouped_avg = data.groupby(['Segment', 'Category'])[['Sales', 'Profit']].mean().reset_index()
grouped_avg
```

```python
data2 = grouped_avg.pivot(index='Segment', columns='Category', values=['Sales', 'Profit'])
data2
```

# " ↑ In this DataFrame, the company can analyze which segments have higher and lower demand."

```python
data3 = data.pivot_table(index="Region",columns="Category",aggfunc="sum",values=["Sales","Profit"])
data3
```

# " ↑ In this DataFrame, the table is organized to visualize which region has the highest and lowest sales and profit gained."
